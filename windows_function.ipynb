```python
from pyspark.sql import SparkSession
from pyspark.sql.window import Window
from pyspark.sql import functions as F

spark = SparkSession.builder.appName("WindowFunctions").getOrCreate()

data = [
    ("A", "2024-01-01", 100),
    ("A", "2024-01-02", 150),
    ("A", "2024-01-03", 200),
    ("B", "2024-01-01", 50),
    ("B", "2024-01-02", 120),
    ("B", "2024-01-02", 120),
    ("B", "2024-01-02", 180),
]
columns = ["Category", "Date", "Sales"]

df = spark.createDataFrame(data, columns)

windowSpec = Window.partitionBy("Category").orderBy("Sales")

# Apply window functions
result = df.withColumn("RowNumber", F.row_number().over(windowSpec)) \
           .withColumn("PrevSales", F.lag("Sales", 1).over(windowSpec)) \
           .withColumn("NextSales", F.lead("Sales", 1).over(windowSpec)) \
           .withColumn("DenseRank", F.dense_rank().over(windowSpec))\
           .withColumn("Rank", F.rank().over(windowSpec))



result.show()
```
