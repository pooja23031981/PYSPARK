```python
from pyspark.sql import SparkSession
from datetime import date

# Start Spark session
spark = SparkSession.builder.appName("EmployeeDataFrame").getOrCreate()

# Employee data (list of tuples)
data = [
    (101, "Alice", "HR", 50000, "2021-01-15"),
    (102, "Bob", "Finance", 60000, "2020-03-10"),
    (103, "Charlie", "IT", 75000, "2022-07-25"),
    (104, "David", "Marketing", 55000, "2019-11-30"),
    (105, "Eva", "IT", 80000, "2021-09-14")
]

# Column names
columns = ["EmpID", "Name", "Department", "Salary", "JoiningDate"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Show DataFrame
df.show()
```

```python
from pyspark.sql import functions as F

df.select(
    F.upper("Name").alias("Upper_Name"),
    F.length("Name").alias("Length"),
    F.substring("Name", 1, 3).alias("First3"),
    F.instr("Name", "a").alias("Pos_of_a")
).show()
```

```python
from pyspark.sql import functions as F

df = spark.createDataFrame([("   Spark   ",)], ["word"])

df.select(
    F.ltrim("word").alias("ltrim"),
    F.rtrim("word").alias("rtrim"),
    F.trim("word").alias("trim"),
    F.lpad("word", 15, "*").alias("lpad"),
    F.rpad("word", 15, "*").alias("rpad")
).show(truncate=False)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import lpad

spark = SparkSession.builder.appName("LPADExample").getOrCreate()

data = [("101",), ("25",), ("7",)]
df = spark.createDataFrame(data, ["id"])

df = df.withColumn("padded_id", lpad("id", 4, "0"))

df.show()
```

```python
from pyspark.sql import functions as f
salary_new=df.groupBy("Department").agg(f.sum("Salary"),f.avg("Salary"),f.max("Salary"),f.min("Salary"),f.count("EmpID"))
salary_new.show()
```

```python
from pyspark.sql import functions as f
salary_new=df.filter(df.Department=="IT").groupBy("Department").agg(f.sum("Salary"),f.avg("Salary"),f.max("Salary"),f.min("Salary"),f.count("EmpID"))
salary_new.show()
```
