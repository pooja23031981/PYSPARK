```python
from pyspark.sql import SparkSession
from datetime import date

# Start Spark session
spark = SparkSession.builder.appName("EmployeeDataFrame").getOrCreate()

# Employee data (list of tuples)
data = [
    (101, "Alice", "HR", 50000, "2021-01-15"),
    (102, "Bob", "Finance", 60000, "2020-03-10"),
    (103, "Charlie", "IT", 75000, "2022-07-25"),
    (104, "David", "Marketing", 55000, "2019-11-30"),
    (105, "Eva", "IT", 80000, "2021-09-14")
]

# Column names
columns = ["EmpID", "Name", "Department", "Salary", "JoiningDate"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Show DataFrame
df.show()
```

```python
from pyspark.sql import functions as F

# Group by Department and calculate salary stats + employee count
salary_stats = df.groupBy("Department").agg(
    F.count("EmpID").alias("EmployeeCount"),
    F.sum("Salary").alias("TotalSalary"),
    F.avg("Salary").alias("AverageSalary"),
    F.max("Salary").alias("MaxSalary"),
    F.min("Salary").alias("MinSalary")
)

salary_stats.show()
```

```python
from pyspark.sql import functions as F

# Filter only IT department before grouping
salary_stats_it = df.filter(df.Department == "IT").groupBy("Department").agg(
    F.count("EmpID"),
    F.sum("Salary"),
    F.avg("Salary"),
    F.max("Salary"),
    F.min("Salary")
)

salary_stats_it.show()
```

```python
df.where(df.Salary > 60000).show()
```

```python
df.where((df.Department == "IT") & (df.Salary > 75000)).show()
```

```python
from pyspark.sql import functions as F

salary_stats = df.groupBy("Department").agg(
    F.count("EmpID").alias("EmployeeCount"),
    F.avg("Salary").alias("AverageSalary")
)

# Apply WHERE condition (average salary > 60000)
salary_stats.where(salary_stats.AverageSalary > 60000).show()
```

```python
from pyspark.sql import functions as F

# Convert string to date type
df_date = df.withColumn("JoiningDate", F.to_date("JoiningDate", "yyyy-MM-dd"))

# Filter employees who joined after 2021-01-01
recent_employees = df_date.where(df_date.JoiningDate > F.lit("2021-01-01"))

recent_employees.show()
```

```python
salary_stats_recent = df_date.where(df_date.JoiningDate > "2021-01-01") \
    .groupBy("Department") \
    .agg(
        F.count("EmpID").alias("EmployeeCount"),
        F.sum("Salary").alias("TotalSalary"),
        F.avg("Salary").alias("AverageSalary")
    )

salary_stats_recent.show()
```

```python
from pyspark.sql import functions as F

# Convert JoiningDate column to proper date type
df_date = df.withColumn("JoiningDate", F.to_date("JoiningDate", "yyyy-MM-dd"))

# Filter employees by date range
employees_between = df_date.where(
    (df_date.JoiningDate >= "2020-01-01") &
    (df_date.JoiningDate <= "2021-12-31")
)

employees_between.show()
```

```python
# Order by Salary descending
employees_between_sorted_salary = employees_between.orderBy(employees_between.Salary.desc())

employees_between_sorted_salary.show()
```

```python
# Order employees by Salary (desc) and take top 2
top2_salary = df.orderBy(df.Salary.desc()).limit(2)

top2_salary.show()
```

```python
# Employees between 2020-01-01 and 2021-12-31 ordered by JoiningDate
employees_between_sorted = employees_between.orderBy("JoiningDate")

employees_between_sorted.show()
```

```python
salary_stats_between_sorted = salary_stats_between.orderBy(F.col("AverageSalary").desc())

salary_stats_between_sorted.show()
```

```python
salary_stats_between_sorted = salary_stats_between.orderBy(F.col("AverageSalary").desc())

salary_stats_between_sorted.show()
```

```python
salary_stats_top = salary_stats_between.orderBy(F.col("AverageSalary").desc()).limit(2)

salary_stats_top.show()
```
