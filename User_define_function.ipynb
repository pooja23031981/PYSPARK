```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType, IntegerType, FloatType
spark = SparkSession.builder.appName("UDF Example").getOrCreate()

data = [("Alice", 20), ("Bob", 35), ("Charlie", 29)]
columns = ["Name", "Age"]

df = spark.createDataFrame(data, columns)
df.show()
def age_group(age):
    if age < 25:
        return "Young"
    elif age < 35:
        return "Adult"
    else:
        return "Senior"

age_group_udf = udf(age_group, StringType())
df = df.withColumn("Age_Group", age_group_udf(df["Age"]))
df.show()

```

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions
from pyspark.sql.functions import udf, col
from pyspark.sql.types import StringType

spark = SparkSession.builder.appName("UDFExample").getOrCreate()

data = [
    ("Laptop", 60000),
    ("Mouse", 300),
    ("Keyboard", 800),
    ("Monitor", 12000)
]

df = spark.createDataFrame(data, ["Product", "Price"])
df.show()
```

```python
def price_category(price):
    if price < 500:
        return "Low"
    elif price <= 1000:
        return "Medium"
    else:
        return "High"
```

```python
category_udf =udf(price_category, StringType())
df = df.withColumn("Category", category_udf(col("Price")))
df.show()
```

```python
add_ten_udf = udf(lambda x: x + 10, IntegerType())
df = df.withColumn("AgePlus10", add_ten_udf(df["Age"]))
df.show()
```
